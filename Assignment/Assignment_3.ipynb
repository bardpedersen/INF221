```
def f1(n):
    result = [1 for _ in range(10)]  #  constant time c*10
    return result
``` 
Here the time complexety is c*10, and can be written as <u>O(1)</u>  

```
def f2(n):
    total = 0
    for i in range(n): # c1*n
        total += i # constant time c2
    return total
```
Here the time complexety is c1*n+c2, and can be written as <u>O(n)</u>

```

    matrix = [[0 for _ in range(n)] for _ in range(n)]  # c1*n * c2*n
```
